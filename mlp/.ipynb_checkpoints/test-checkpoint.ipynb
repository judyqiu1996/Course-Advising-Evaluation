```python
import csv
import numpy as np
```

```python
data1 = open("homework.csv")
data2 = open("mid1.csv")
data3 = open("mid2.csv")
data4 = open("final.csv")
```

```python
row1 = data1.readlines()
row2 = data2.readlines()
row3 = data3.readlines()
row4 = data4.readlines()
```

```python
hw,mid1,mid2,final = [],[],[],[]
for i,j,k,l in zip(row1,row2,row3,row4):
    hw.append(i.split(","))
    mid1.append(j.split(","))
    mid2.append(k.split(","))
    final.append(l.split(","))
```

```python
for i in hw:
    print(i)
```

```python
num = len(hw)-1
num
```

```python
homework = np.zeros(num)
count =0
for i in range(len(hw[0])):
    if hw[0][i] == "hm_score":
        count+=1
        for j in range(1,len(hw)):
            #print(hw[j][i])
            if hw[j][i] == '' or '#N/A':
                hw[j][i]=0
            #print(hw[j][i])
            homework[j-1] += float(hw[j][i])
            #print(hw[j][i])
homework /= count
```
